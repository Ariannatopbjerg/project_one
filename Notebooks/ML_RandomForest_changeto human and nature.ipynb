```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.svm import SVC
import pandas as pd
import tensorflow as tf

# Import our input dataset
fire_df = pd.read_csv('../Resources/wildfire_v2_clean_data.csv')
fire_df.head()
```

```python
fire_clean_df=fire_df.drop(columns=["discovery_date","fire_size","fire_mag","state","discovery_month","discovery_year"])
fire_clean_df
```

```python
fire_cause_unique=fire_clean_df.fire_size_class.value_counts()
fire_cause_unique
```

```python
devastating = ["E","F","G"]
not_devastating = ["B","C","D"]

# Replace in dataframe
for app in devastating:
    fire_clean_df.fire_size_class = fire_clean_df.fire_size_class.replace(app,"devastating")
for app in not_devastating:
    fire_clean_df.fire_size_class = fire_clean_df.fire_size_class.replace(app,"minor")
```

```python
fire_clean_df
```

```python
feature_df=fire_clean_df.drop(columns=["fire_size_class"])
```

```python
# Generate our categorical variable list
fire_cat = feature_df.dtypes[feature_df.dtypes == "object"].index.tolist()

# Check the number of unique values in each column
feature_df[fire_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(feature_df[fire_cat]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(fire_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
x_df = feature_df.merge(encode_df,left_index=True, right_index=True)
x_df = x_df.drop(fire_cat,1)
x_df.head()
```

```python
# Remove target from features data
y = fire_clean_df["fire_size_class"].tolist()
X = x_df
# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_test
```

```python

from sklearn.ensemble import RandomForestClassifier
# Create a random forest Classifier.
rf_model = RandomForestClassifier(n_estimators=100, random_state=2)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
